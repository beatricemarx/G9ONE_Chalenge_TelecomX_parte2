```python
#Teste do Q quadrado - Análise de variáveis categóricas: Cria um rank das Variáveis que mais tem relação de responder sim para o churn



from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import chi2

# Criar cópia
df_rank = df.copy()

# Converter Churn para binário
df_rank["Churn"] = df_rank["Churn"].map({"Yes": 1, "No": 0})

# Separar variáveis categóricas
categorical_cols = df_rank.select_dtypes(include=["object"]).columns

# Aplicar Label Encoding temporário (apenas para teste estatístico)
le = LabelEncoder()
for col in categorical_cols:
    df_rank[col] = le.fit_transform(df_rank[col])

# Separar X e y
X_cat = df_rank[categorical_cols]
y = df_rank["Churn"]

# Aplicar teste Qui-quadrado
chi_scores, p_values = chi2(X_cat, y)

# Criar DataFrame com ranking
chi_df = pd.DataFrame({
    "Variável": categorical_cols,
    "Chi2 Score": chi_scores,
    "p-value": p_values
})

# Ordenar por importância
chi_df = chi_df.sort_values(by="Chi2 Score", ascending=False)

chi_df


#Quanto maior o Chi2 Score maior a relação
#p-value - quanto menor o valor mais confiável é o chi2 score
```

```python
significant_vars = chi_df[chi_df["p-value"] < 0.05]
print("\n✅ Variables significativas (p < 0.05):")
display(significant_vars)
```

## 🎯 Split

1. Separação de X e Y

```python
X = df.drop("Churn", axis=1)
y = df["Churn"]
```

2. Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

3.One-Hot Encoding

```python
# Aplicar One-Hot Encoding no treino
X_train = pd.get_dummies(X_train, drop_first=True)

# Aplicar no teste
X_test = pd.get_dummies(X_test, drop_first=True)

# Garantir que treino e teste tenham as mesmas colunas
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
```

```python
X_train.head()
```

4. Verificar desbalaceamento

```python
y_train.value_counts(normalize=True) * 100
```

##Treinamento

Modelo 1 — Regressão Logística

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)

y_pred_log_normal = log_model.predict(X_test)

print("Regressão Logística")
print(classification_report(y_test, y_pred_log_normal))
```

Modelo 2 — Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

print("Random Forest")
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000, class_weight="balanced")
log_model.fit(X_train, y_train)

y_pred_log = log_model.predict(X_test)

print(classification_report(y_test, y_pred_log))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import numpy as np

# Matriz de confusão
cm = confusion_matrix(y_test, y_pred_log)

# Plot
plt.figure()
plt.imshow(cm)
plt.title("Matriz de Confusão - Regressão Logística (Balanceada)")
plt.xlabel("Predito")
plt.ylabel("Real")

# Colocar números dentro da matriz
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.xticks([0,1], ["No", "Yes"])
plt.yticks([0,1], ["No", "Yes"])

plt.show()
```

# Relatório Final

Este relatório detalha a análise de evasão de clientes (Churn), abrangendo o processamento de dados, os modelos de machine learning utilizados e as estratégias sugeridas para retenção.


**1. Principais Fatores que Influenciam a Evasão**

Com base na análise exploratória e no desempenho das variáveis nos modelos, os fatores que mais impactam a decisão de saída do cliente são:

**Tipo de Contrato:** Clientes com contratos de mês a mês (Month-to-month) apresentam uma taxa de evasão significativamente maior do que aqueles com contratos anuais ou bienais.

**Serviço de Internet:** O uso de Fibra Óptica (Fiber optic) está correlacionado a uma maior taxa de churn, possivelmente devido a preços mais elevados ou instabilidades percebidas no serviço em comparação ao DSL.

**Tempo de Permanência (Tenure):** Clientes novos (com baixo tenure) são muito mais propensos a cancelar o serviço. A probabilidade de evasão diminui à medida que o tempo de relacionamento com a empresa aumenta.

**Faturamento Sem Papel e Método de Pagamento:** O uso de Paperless Billing e pagamentos via Electronic Check foram identificados como variáveis frequentes em perfis de clientes que cancelam o serviço.

**Serviços Adicionais:** A ausência de Suporte Técnico (TechSupport) e Segurança Online (OnlineSecurity) aumenta drasticamente as chances de evasão.

**2. Desempenho dos Modelos**

Foram testados diferentes modelos de classificação para prever o churn. O desempenho é geralmente avaliado por métricas como Acurácia, Recall e F1-Score:

Regressão Logística: Frequentemente utilizada como base (baseline), oferece boa interpretabilidade sobre o peso de cada variável.

Random Forest / Modelos de Árvore: Tendem a apresentar melhor desempenho ao capturar relações não lineares entre as variáveis, como o impacto combinado de altos custos mensais com contratos curtos.

K-Nearest Neighbors (KNN): Também aplicado para identificar similaridades entre perfis de clientes evadidos e ativos.

**3. Estratégias de Retenção Propostas**

Com base nos resultados, as seguintes estratégias de negócio são recomendadas:

Incentivo à Fidelidade: Criar campanhas para migrar clientes de contratos mensais para contratos de 1 ou 2 anos, oferecendo descontos progressivos ou benefícios exclusivos.

Foco no Suporte Técnico: Oferecer pacotes de TechSupport e OnlineSecurity gratuitos por um período determinado para clientes de Fibra Óptica, visando reduzir a percepção de problemas técnicos.

Revisão do Processo de Cobrança: Analisar por que clientes que utilizam boleto eletrônico (Electronic Check) evadem mais e oferecer alternativas como débito automático com benefícios.
